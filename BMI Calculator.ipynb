# BMI (Body_Mass_Index) Calculator

# BMI Formula : BMI = mass (kg)/height^2 (m)

```python
name = input("Enter Your Name")
weight = int(input("Enter Your Weight In Kg: ")) 
height_in_m = int(input("Enter Your Height In feet: "))
height_in_inches = int(input("Enter Your Height In inches: "))

height = (height_in_m * 0.3048) + (height_in_inches * 0.0254)

BMI = weight/height**2

print("Your BMI is : " , BMI)
```

**Classification	BMI range - kg/m2
Severe Thinness	< 16
Moderate Thinness	16 - 17
Mild Thinness	17 - 18.5
Normal	18.5 - 25
Overweight	25 - 30
Obese Class I	30 - 35
Obese Class II	35 - 40
Obese Class III	> 40**

```python
if(BMI < 16):
    print("Severe Thinness")
elif(BMI > 16 and BMI < 17):
    print("Moderate Thinness")
elif(BMI > 17  and BMI < 18.5):
    print("Mild Thinness")
elif(BMI > 18.5 and BMI < 25):
    print("Normal")
elif(BMI > 25 and BMI < 30):
    print("Overweight")
else : 
    print("Obese")
```
