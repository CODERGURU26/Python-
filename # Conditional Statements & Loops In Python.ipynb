# Conditional Statements & Loops In Python


# 1. if-else Statements

#if statement

age = 20
if(age > 18):
    print('Adult')

```python
#if-else statement

age = 14
if(age > 18):
    print('Eligble For Voting')
else:
    print('Not Eligible For Voting')
```

```python
#elif statement

marks = 60
if(marks >= 75 and marks <= 90):
    print('Grade A')
elif(marks >=60 and marks <= 75):
    print('Grade B')
else:
    print('Grade C')
```

# 2. For loop

```python
# lists using for loop
number = [1 , 2 , 3 , 4 , 5]
for i in number:
    print(i)
```

```python
#dictionaries using for loop
dict_info = {'name' : 'Gururaj' , 'age': 20 , 'gender': 'male'} 
for i in dict_info.values():
    print(i)
```

```python
#nested for loops
cars = ['BMW' , 'Buggati' , 'Mustang' , 'ferrari' , 'lamborgini']
bikes = ['Ducati' , 'Hayabuza' , 'Royal Enfield' , 'Harley Davidson' , 'Kawasaki Ninja']

for i in cars:
    for j in bikes:
        print('Cars:' , i  , 'and bikes : ' , j) 
```

# 3. While Loop


```python
number = 0

while(number <= 5):
    print(number)
    number = number + 1
```

```python
#break in while loop
number = 0

while(number <= 5):
    print(number)
    if(number == 3):
        break
    number = number + 1
```

```python
#continue in while loop
number = 0

while(number < 5):
    number = number + 1

    if(number == 3):
        continue

    print(number)
# it will skip the number 3
```
