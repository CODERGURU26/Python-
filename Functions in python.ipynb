# Functions In Python

```python
def my_name(name):
    print('My Name: ' , name)
```

```python
my_name('Gururaj')
```

```python
def power(number , power):
    print(number ** power)
```

```python
power(3 , 4)
```

```python
#arbitary arguments
def arb_arguments(*number):
    print(number[0] + number[1])
```

```python
arb_arguments(4 , 5 ,6 , 7 ,  8, 9) 
```

```python
arg_tuples = (4 , 5 ,6 , 7 ,  8, 9)

def arb_arguments(*number):
    print(number[0] + number[1])
```

```python
arb_arguments(*arg_tuples) 
```

```python
#keyword arguments
def power(number , power):
    print(number ** power)
```

```python
power(power = 2 , number = 10)
```
