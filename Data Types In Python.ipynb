# Data Types In Python

```python
#integer

x = 10
type(x)
```

```python
#float
y = 4.5
type(y)
```

```python
#boolean
z = 1 > 5
type(z)
```

```python
#strings
str = "Hello World"
type(str)
```

```python
#indexing
print(str[:5])
```

```python
#lists - They are mutable:means you can edit them
ice_cream = ['Chocolate' , 'Vanilla' , 'Butterscotch']
type(ice_cream)
```

```python
#printing list
print(ice_cream)
```

```python
# multiple datatypes in single list : nested_list
nested_list = ['Vanilla' , 3 , ['Chocolate' , 'Choco'] , True]
print(nested_list)
```

```python
#append 
ice_cream.append('Caramel') 
```

```python
#indexing in list 
ice_cream[0] = 'ChocoChip'
```

```python
#tuples - they are immutable : means u cant edit them 
tuple_scoops = (1 , 2 , 3 , 4 , 5)
type(tuple_scoops)
```

```python
tuple_scoops[0]
```

```python
#sets 
set1 = {1 , 2 , 3 , 4 , 5 }
set2 = {5 , 2 , 7 , 8 , 9 }
print(set1 , set2)
```

```python
#Union - Return Unique elements from sets
print(set1 | set2)
```

```python
#Intersection - Return Common Elements from sets
print(set1 & set2)
```

```python
#Difference - gives you elements that are in set1 but NOT in set2
print(set1 - set2)
```

```python
#symmetric difference - gives you elements that are in either set1 or set2 but NOT in both.
print(set1 ^ set2)
```

```python
#Dictionaries - Key:Value pair
info_dict = {'name' : 'Gururaj' , 'age' : 20 , 'gender' : 'male' }
print(info_dict)
```

```python
info_dict.keys()
```

```python
info_dict.values()
```

```python
info_dict.items()
```

```python
#accessing elements in dictionaries
info_dict['name']
```

```python
#updating dictionaries
info_dict.update({'name' : 'Rohan'})
```

```python
#deleting in dictionaries
del info_dict['gender']
```
